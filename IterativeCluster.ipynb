```python
import matplotlib.pyplot as plt


def get_metrics(y1,y2):
    from sklearn import metrics
    ARI = metrics.adjusted_rand_score(y1,y2)
    AMI=metrics.adjusted_mutual_info_score(y1,y2)
    Homogenity=metrics.homogeneity_score(y1,y2)
    Completeness=metrics.completeness_score(y1,y2)
    V_measure=metrics.v_measure_score(y1,y2)
    Silhouette=metrics.v_measure_score(y1,y2)
    return ("ARI",ARI,"AMI",AMI,"Homogenity",Homogenity,
            "Completeness",Completeness,"V_measure",V_measure,
            "Silhouette",Silhouette)

def model_visualization(y_pred,y, labels,plt,m,n):
    plt.subplot(320+n)
    plt.scatter(X[:,0],X[:,1],c=y_pred,marker='o',s=120,edgecolors='k')
    plt.title("Пример "+str(n)+"\n"+m[0]+"="+str(m[1]))

from sklearn.cluster import KMeans, affinity_propagation
import matplotlib.pyplot as pl
import numpy as np

plt.figure(figsize=(12,16))

from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=200,random_state=150)
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,1)

from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=200,random_state=150)
transformation=[[0.61,-0.64],[-0.4,0.8]]
X=np.dot(X,transformation)
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,2)

from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=200,cluster_std=[1.0,2.5,0.5],random_state=150)
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,3)

from sklearn.datasets import make_blobs
X,y=make_blobs(n_samples=1000,random_state=150)
X=np.vstack((X[y==0][:150],X[y==1][:40],X[y==2][:10]))
y=np.concatenate([np.tile(0,150),np.tile(1,40),np.tile(2,10)])
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,4)

from sklearn.datasets import make_circles
X,y=make_circles(n_samples=200,factor=.5,noise=.05)
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,5)

from sklearn.datasets import make_moons
X,y=make_moons(n_samples=200,noise=.05)
model=KMeans(n_clusters=3,random_state=0)
model.fit(X)
y_pred=model.predict(X)
m=get_metrics(y,model.labels_)
model_visualization(y_pred,y,model.labels_,plt,m,6)

plt.show()
```
